```python
import random
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
x = []
y = []
```

```python
for i in range(10):
  x.append(random.randint(0,50))
  y.append(random.randint(0,50))
```

```python
print(x)
print(y)
```

```python
plt.scatter(x, y, color='blue', label='Data')
plt.legend()
plt.show()
```

```python
m, c = np.polyfit(x,y,1)  #a method that fits the data within a polynomial function with degree 1.
```

```python
x1 = np.linspace(min(x),max(x))
```

```python
y1 = m * x1 + c
```

```python
print(len(x1))
len(y1)
```
