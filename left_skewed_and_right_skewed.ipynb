```python
import matplotlib.pyplot as plt
import numpy as np

# Assuming you already have the simulated lifespan data in the 'lifespans' variable

# Fit a normal distribution to the data (optional)
from scipy.stats import norm

# Replace 'lifespans' with your actual data if you didn't perform the fit
mu, sigma = norm.fit(lifespans)  # Estimate mean (mu) and standard deviation (sigma)

# Create x-axis values covering the data range with finer resolution
x = np.linspace(min(lifespans), max(lifespans), 1000)

# Calculate the probability density function (PDF) of the normal distribution at each x-axis value
y = norm.pdf(x, mu, sigma)  # mu is mean, sigma is standard deviation

# Plot the histogram (already created) and the normal distribution curve
plt.hist(lifespans, density=True, alpha=0.5, label='Histogram')  # Adjust alpha for transparency
plt.plot(x, y, label='Normal Distribution')

# Customize the plot for clarity
plt.xlabel("Age at Death (Years)")
plt.ylabel("Probability Density")
plt.title("Simulated Lifespan Distribution")
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import lognorm  # Requires scipy library

# Simulate left-skewed data (adjust parameters as needed)
try:
  # Try using 'shape' parameter for newer scipy versions
  data = np.random.lognormal(50, size=1000)  # Controls central tendency
except TypeError:
  # If 'shape' fails, use 'scale' for older versions (fallback)
  data = np.random.lognormal(scale=50, size=1000)

# Create x-axis for the estimated normal distribution curve (finer resolution)
x = np.linspace(min(data), max(data), 1000)

# Generate estimated normal distribution curve (might not perfectly fit the left skew)
estimated_mu = np.mean(data)
estimated_sigma = np.std(data)
y = norm.pdf(x, estimated_mu, estimated_sigma)  # Probability density function

# Create the histogram
plt.hist(data, density=True, bins=50, alpha=0.7, label='Left-Skewed Distribution')  # Adjust 'bins' for bar count

# Plot the estimated normal distribution curve
plt.plot(x, y, label='Estimated Normal Distribution (for visualization)')

# Customize the plot for clarity
plt.xlabel("Values")
plt.ylabel("Probability Density")
plt.title("Left-Skewed Distribution with Estimated Normal Curve")
plt.legend()
plt.grid(True)  # Add gridlines for better readability
plt.show()
```
